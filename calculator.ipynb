```python
history = []
operators = ["+", "-", "*", "/"]
is_running = True

def calculate(a, op, b):
    a = float(a)
    b = float(b)
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b == 0:
            return "Error: 0으로 나눌 수 없습니다"
        return a / b
    else:
        return "Error: 지원하지 않는 연산자"

while is_running:
    a = input("첫 번째 숫자: ")
    if a == "q":
        print("종료합니다.")
        break
    op = input("연산자: ")
    b = input("두 번째 숫자: ")
    result = calculate(a, op, b)
    print("결과:", result)
    history.append(str(a) + op + str(b) + "=" + str(result))

for record in history:
    print(record)
```
