```python
# ==========================================
# 변수 선언 (문장 12)
# ==========================================
history = []
operators = {"+": "더하기", "-": "빼기", "*": "곱하기", "/": "나누기"}
is_running = True


# ==========================================
# 함수 1 — calculate() (문장 1~6)
# ==========================================
def calculate(a, op, b):
    a = float(a)                                        # 문장 2
    b = float(b)

    if op == "+":                                       # 문장 3
        result = a + b
    elif op == "-":
        result = a - b
    elif op == "*":
        result = a * b
    elif op == "/":
        if b == 0:                                      # 문장 4
            return "Error: 0으로 나눌 수 없습니다", False
        result = a / b
    else:                                               # 문장 5
        return "Error: 지원하지 않는 연산자입니다", False

    return result, True                                 # 문장 6


# ==========================================
# 함수 2 — save_history() (문장 7~8)
# ==========================================
def save_history(a, op, b, result):
    record = str(a) + " " + op + " " + str(b) + " = " + str(result)
    history.append(record)


# ==========================================
# 함수 3 — show_history() (문장 9~11)
# ==========================================
def show_history():
    if len(history) == 0:
        print("계산 기록이 없습니다.")
        return
    print("=== 계산 기록 ===")
    for i, record in enumerate(history):
        print(str(i + 1) + ". " + record)


# ==========================================
# 메인 실행부 — while 루프 (문장 13~16)
# ==========================================
print("=== 파이썬 계산기 ===")
print("종료하려면 q를 입력하세요")

while is_running:

    a = input("첫 번째 숫자: ")

    if a == "q":                                        # 문장 14
        print("계산기를 종료합니다.")
        break

    op = input("연산자 (+,-,*,/): ")
    b = input("두 번째 숫자: ")

    result, success = calculate(a, op, b)              # 문장 15

    if success:
        print("결과:", result)
        save_history(a, op, b, result)
    else:
        print(result)

    view = input("히스토리 볼까요? (y/n): ")           # 문장 16
    if view == "y":
        show_history()


```
